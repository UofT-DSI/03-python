```python
def anagram_checker(word_a,word_b,is_case_sensative=False):
    "Turns two word into list and compare the sets"
    if is_case_sensative == False:
        a = set(word_a.lower()) 
        b = set(word_b.lower())
        return a == b

    elif is_case_sensative == True:
            A = set(word_a)
            B = set(word_b)
            return A == B
         
```

```python
anagram_checker("Slient", "Night")
```

```python
anagram_checker("night", "Thing")
```

```python
anagram_checker("Slient", "listen")
```

```python
anagram_checker("Slient", "listen", False)
```

```python
anagram_checker("Slient", "Listen", True)
```
