```python
# This is part two 

def anagram_checker(word_a,word_b,is_case_sensative=False):
    "Turns two word into list and compare the sets"
    if not is_case_sensative:
        word_a = sorted(word_a.lower())
        word_b = sorted(word_b.lower())
        return word_a == word_b

    else:
            A = sorted(word_a)
            B = sorted(word_b)
            return A == B
         
```

```python
anagram_checker("Slient", "listen", False)
```

```python
anagram_checker("Slient", "listen", True)
```

```python
anagram_checker("list","List")
```

```python
anagram_checker("L", "LL", True)
```
